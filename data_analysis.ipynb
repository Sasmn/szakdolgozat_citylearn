```python
# System operations
import os
from citylearn.data import DataSet
import matplotlib.pyplot as plt
import pandas as pd
import citylearn
```

```python
# set all plotted figures without margins
plt.rcParams['axes.xmargin'] = 0
plt.rcParams['axes.ymargin'] = 0

print(DataSet.get_names())
print(citylearn.__version__)
```

```python
DATASET_NAME = 'citylearn_challenge_2022_phase_all'
schema = DataSet.get_schema(DATASET_NAME)
```

```python
root_directory = schema['root_directory']

building_1 = 'Building_1'
building_2 = 'Building_2'
building_3 = 'Building_3'
building_4 = 'Building_4'
building_5 = 'Building_5'
building_6 = 'Building_6'
```

```python
# ENERGY SIMULATION CSV
filename = schema['buildings'][building_1]['energy_simulation']
filepath = os.path.join(root_directory, filename)
building_data = pd.read_csv(filepath)
display(building_data.head())

display(building_data.describe(include='all'))

%matplotlib inline
fig, axs = plt.subplots(1, 2, figsize=(18, 2))
x = building_data.index
y1 = building_data['Equipment Electric Power [kWh]']
y2 = building_data['Solar Generation [W/kW]']
axs[0].plot(x, y1)
axs[0].set_xlabel('Time step')
axs[0].set_ylabel('Equipment Electric Power\n[kWh]')
axs[1].plot(x, y2)
axs[1].set_xlabel('Time step')
axs[1].set_ylabel('Solar Generation\n[W/kW]')
fig.suptitle(building_1)
plt.show()
```

```python
import numpy as np
filename = schema['buildings'][building_1]['energy_simulation']
filepath = os.path.join(root_directory, filename)
building_data = pd.read_csv(filepath)

solar = building_data["Solar Generation [W/kW]"]

last_end_index = 0;

numberOfSubCategories = 12
fig, axs = plt.subplots(4, 3, figsize=(10, 10))


for ax, i in zip(fig.axes, range(numberOfSubCategories)):
    start_index = last_end_index
    end_index = int(8760 / numberOfSubCategories * i + 8760 / numberOfSubCategories)
    for j in range(start_index, end_index, 24):
        ax.plot(solar[j:j+24].values, alpha=0.9, color="gray", lw=0.5)
        last_end_index = j + 24;

    filtered_data = building_data[(building_data.index >= start_index) & (building_data.index < end_index)]
    quantiles = np.arange(0.1, 1.1, 0.1)
    colors = plt.cm.viridis(np.linspace(0, 1, len(quantiles)))
    # ax.plot(filtered_data.groupby(["Hour"])["Solar Generation [W/kW]"].quantile(0.5), color="black", label="Median Production in M" + str(i + 1))
    for q, color in zip(quantiles, colors):
        quantile_data = filtered_data.groupby("Hour")["Solar Generation [W/kW]"].quantile(q)
        ax.plot(quantile_data, color=color)
    ax.set_ylabel("Solar Generation")
    ax.set_xlabel("Time of Day")
    ax.set_ylim(solar.min(), solar.max())
    ax.set_title("M" + str(i + 1))

plt.tight_layout()
```

```python
# WHEATHER CSV

filename = schema['buildings'][building_1]['weather']
filepath = os.path.join(root_directory, filename)
weather_data = pd.read_csv(filepath)
display(weather_data.head())

display(weather_data.describe(include='all'))

%matplotlib inline
columns = [
    'Outdoor Drybulb Temperature [C]', 'Relative Humidity [%]',
    'Diffuse Solar Radiation [W/m2]', 'Direct Solar Radiation [W/m2]'
]
fig, axs = plt.subplots(2, 2, figsize=(18, 5))
x = weather_data.index

for ax, c in zip(fig.axes, columns):
    y = weather_data[c]
    ax.plot(x, y)
    ax.set_xlabel('Time step')
    ax.set_ylabel(c)

fig.align_ylabels()
plt.tight_layout()
plt.show()
```
