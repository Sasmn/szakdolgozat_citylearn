```python
import pandas as pd

csv_file_path = 'weather_data_budapest_xi.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path, sep=';\\s*', header=0)
```

```python
# Convert the 'Datetime' column to datetime format
df['Datetime'] = pd.to_datetime(df['Time'], format='%Y%m%d%H%M', errors='coerce')

# # Filter the DataFrame for the specified date range (May to September 2023)
filtered_df = df[(df['Datetime'].dt.year == 2023) & (df['Datetime'].dt.month >= 5) & (df['Datetime'].dt.month <= 9)]

# print(filtered_df)

# Initialize a dictionary to store data month by month
budapest_weather_data = []

# Group the filtered DataFrame by month and iterate through each group
for name, group in filtered_df.groupby(filtered_df['Datetime'].dt.month):
    # Extract month and year for labeling
    month_year = group['Datetime'].dt.strftime('%B %Y').iloc[0]
    
    # Store the relevant data (Datetime, Outdoor drybulb temperature, Outdoor relative humidity) for each month
    month_data = {
        'Month': group['Datetime'].dt.month,
        'Day': group['Datetime'].dt.day,
        'Hour': group['Datetime'].dt.hour,
        'Outdoor Drybulb Temperature (C)': group['t'],
        'Outdoor Relative Humidity (%)': group['u']
    }

    budapest_weather_data.append(pd.DataFrame(month_data))

budapest_weather_data = pd.concat(budapest_weather_data, ignore_index=True)

display(budapest_weather_data.head())
```

```python
num_days = [30, 31, 31]
num_hours = 24
months = [6, 7, 8]

hours = []
days = []
months_list = []

for month, days_in_month in zip(months, num_days):
    for day in range(1, days_in_month + 1):
        for hour in range(num_hours):
            hours.append(hour)
            days.append(day)
            months_list.append(month)

# Adding the new properties to the DataFrame
weather_data['Month'] = months_list
weather_data['Day'] = days
weather_data['Hour'] = hours

# Reordering the columns
weather_data = weather_data[['Month', 'Day', 'Hour'] + [col for col in weather_data.columns if col not in ['Month', 'Day', 'Hour']]]

display(weather_data.head())
```

```python
mean_temp_budapest = budapest_weather_data.groupby('Month')['Outdoor Drybulb Temperature (C)'].mean()
mean_temp_citylearn = weather_data.groupby('Month')['Outdoor Drybulb Temperature (C)'].mean()

# Find the month in data2 that matches each month in data1 the most
best_matches = {}
for month_budapest, temp_budapest in mean_temp_budapest.items():
    best_match = None
    min_difference = float('inf')
    for month_citylearn, temp_citylearn in mean_temp_citylearn.items():
        difference = abs(temp_budapest - temp_citylearn)
        if difference < min_difference:
            min_difference = difference
            best_match = month_citylearn
    best_matches[month_budapest] = best_match

# Display the best matches
for month_budapest, month_citylearn in best_matches.items():
    print(f"Month in Budapest weather data (May-Sept): {month_budapest} matches Month in Citylearn weather data (June-Aug): {month_citylearn} the most.")
```

```python
best_matches = {}
best_differences = {}

for month_budapest in range(5, 10):
    budapest_month_data = budapest_weather_data[budapest_weather_data['Month'] == month_budapest]
    
    best_match_month = None
    min_difference = float('inf')
    
    for month_citylearn in range(6, 9):
        citylearn_month_data = weather_data[weather_data['Month'] == month_citylearn]
        
        total_difference = 0
        for day in range(1, 30):
            
            for hour in range(24):
                temp_budapest_hour = budapest_month_data[(budapest_month_data['Hour'] == hour) & (budapest_month_data['Day'] == day)]['Outdoor Drybulb Temperature (C)'].values[0]
                temp_citylearn_hour = citylearn_month_data[(citylearn_month_data['Hour'] == hour) & (citylearn_month_data['Day'] == day)]['Outdoor Drybulb Temperature (C)'].values[0]
                
                difference = abs(temp_budapest_hour - temp_citylearn_hour)
                total_difference += difference
            
        if total_difference < min_difference:
            min_difference = total_difference
            best_match_month = month_citylearn
        
        best_matches[month_budapest] = best_match_month
        best_differences[month_budapest] = min_difference / (30 * 24)

# Display the best matches
for month_budapest, month_citylearn in best_matches.items():
    print(f"Month in Budapest (May-Sept): {month_budapest} matches Month in Citylearn (June-Aug): {month_citylearn} the most, with difference: {best_differences[month_budapest]}")
```

```python
from sklearn.metrics import mean_squared_error

best_matches = {}
best_differences = {}

for month_budapest in range(5, 10):
    budapest_month_data = budapest_weather_data[budapest_weather_data['Month'] == month_budapest]
    
    best_match_month = None
    min_difference = float('inf')
    
    for month_citylearn in range(6, 9):
        citylearn_month_data = weather_data[weather_data['Month'] == month_citylearn]
            
        total_difference = np.sqrt(mean_squared_error(budapest_month_data['Outdoor Drybulb Temperature (C)'][:720], citylearn_month_data['Outdoor Drybulb Temperature (C)'][:720]))
        print(f"Month in Budapest: {month_budapest}, Month in Citylearn: {month_citylearn}, difference: {total_difference}")
        if total_difference < min_difference:
            min_difference = total_difference
            best_match_month = month_citylearn
        
        best_matches[month_budapest] = best_match_month
        best_differences[month_budapest] = min_difference

# Display the best matches
for month_budapest, month_citylearn in best_matches.items():
    print(f"Month in Budapest (May-Sept): {month_budapest} matches Month in Citylearn (June-Aug): {month_citylearn} the most, with difference: {best_differences[month_budapest]}")
```
